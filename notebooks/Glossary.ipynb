## Glossary

### Data Mining 
Using Machine Learning or statistical techniques to explore data in order to
discover patterns.

### Supervised Learning
Training data is _labelled_ with the target values. If the label is a class this
is a classification problem, if some  continuous variable it is a regression
problem. 


### Unsupervised Learning
Training data is _unlabelled_. The model tries to learn by identifying patterns
in the data.

### Semisupervised Learning
Systems that combine supervised and unsupervised techniques to learn from
partially trained data. An example might be  using a supervised model to label
data clusters that have been generated by an unsupervised model.

### Reinforcement Learning
A system that learns a task by reacting to some reward or punishment that is
produced as a result of its current performance. In  reinforcement learning a
system must decide a _policy_ that maximises its _reward_. The system trains
by performing a combination of its existing policy and some new behaviour that
could be random or the product of some other Maching Learning model. If the rate
of reward increases, the policy is updated and the system improves.

### Batch Learning
A system that is trained completely prior to use. A batch, or offline system is
does not improve once it has been trained. In order to adapt to new data, it must
be retrained. 
    

### Online Learning
A system that is able to learn _on the fly_ once it has been deployed. This is 
either done by updating the system sequentially as new training data arrives or
in _mini-batches_. Online systesms can be quicker to adapt and can uses fewer
resources.

### Out-of-core Learning
A system that uses online techniques like _mini-batching_ to process datasets that
may not be feasible to process in a single batch due to resource constraints. For
example if the training dataset cannot fit in memory.


### Instance Based Learning
New cases are compared to a set of existing data. Results are determined by some 
similarity measure. For example, assigning the value to a new datapoint of the most
similar (or the average of k-most similar) previous datapoints.

### Model Based Learning
A model is fit to existing data. New results are predicted by this model.

### Overfitting
A situation where a model performs well on a test data set but the model does not 
generalise well to new data. This occurs when the model is too complex for the
quantity of training data. 

### Underfitting
A model can be too simple to fit the underlying patterns in the data and not perform
well. More complex models or better feature selection can improve the performance.

### Hyperparameter
A hyperparameter is a parameter to a machine learning algorithm, that is selected before
the training process. In order to select the best value for a hyperparameter, the model
can be trained for a range of values, and the best performing selected.



### Precision
The precision of a model is the ratio of true positives to the 
total number of positives predicted (true plus false). This is a
measure of the accuracy of the true predictions. How many of the 
true predictions are correct?


### Recall
The recall of a model, also known as sensitivity is a measure of 
the true postives to total number of positives (predicted or not).
This is a measure of the models ability to identify the true 
samples. How many of the positive cases are identified?

### Precision / Recall Tradeoff
This is a term for the antagonism between these two model metrics.
Perfect precision can be achieved by making a single positive prediction,
but this would then having very low recall. Perfect recall could be 
achieved by marking all samples positive but this would then have
very low precision. Ideally a model has a good score for both values
and in training a model it's important to consider both, and which is
most important for a project.

### F1 Score
A single metric that is the Harmonic Mean of Precision and Recall. The
harmonic mean is used so that low values for either score product a low
value for the total. This has the advantage of a single score with which
to compare models but it is important to understand, in the context of a
project, the importance of both precision and recall.

### Confusion Matrix
Confusion matrices provide a good way to visualise the performance of a
classifier. The columns of the matrix contain the actual classes of the
samples while the rows contain the predictions. Each cell (i, j) then
contains the number of samples of actual class i, with prediction j. The
diagonals of the confusion matrix contain the correctly classified samples. 
A confusion matrix can be used to identify where a classifier is making 
errors.
